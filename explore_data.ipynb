```python
import itertools
import matplotlib.pyplot
import pandas
import sklearn.manifold
```

```python
# Read in data
train_data_url = 'http://download.tensorflow.org/data/iris_training.csv'
train_data_raw = pandas.read_csv(train_data_url)
train_data_raw.columns = ['petal_length', 'petal_width',
                          'sepal_length', 'sepal_width',
                          'class']
```

```python
matplotlib.pyplot.figure(figsize=(18,18))
feature_pairs = list(itertools.product(range(4), range(4)))
for i in range(len(feature_pairs)):
    pair = feature_pairs[i]
    ax = matplotlib.pyplot.subplot(4, 4, i+1)
    matplotlib.pyplot.scatter(train_data_raw.iloc[:, pair[1]], train_data_raw.iloc[:, pair[0]],
                              c=train_data_raw['class'])
    ax.set_xlabel(train_data_raw.columns[pair[1]])
    ax.set_ylabel(train_data_raw.columns[pair[0]])
matplotlib.pyplot.show()
```

```python
train_data_embedded = sklearn.manifold.TSNE().fit_transform(train_data_raw)
matplotlib.pyplot.scatter(train_data_embedded[:, 0], train_data_embedded[:, 1],
                          c=train_data_raw['class'])
matplotlib.pyplot.show()
```
